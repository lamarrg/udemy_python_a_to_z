```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style("dark", {"axes.facecolor":"black"}) #standard -> white, whitegrid, dark, darkgrid, ticks
f, axes = plt.subplots(2, 2, figsize=(15, 15))

#Plot [0,0]
kde1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, \
                   shade=True, shade_lowest=True, cmap='inferno', \
                   ax=axes[0,0])
kde1b = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, \
                    cmap='cool', ax=axes[0,0])

#Plot [0,1]
kde2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, \
                   shade=True, shade_lowest=True, cmap='inferno', \
                   ax=axes[0,1])
kde2b = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, \
                    cmap='cool', ax=axes[0,1])

#Plot [1,0]
kde3 = sns.violinplot(data=movies, \
                      x='Year', y='BudgetMillions', ax=axes[1,0], \
                    palette='YlOrRd')

#Plot[1,1]
kde4 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                shade=True, shade_lowest=False, cmap='Blues_r', ax=axes[1,1])
kde4b = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
             cmap='gist_gray_r', ax=axes[1,1])
kde1.set(xlim=(-30, 160))
kde2.set(xlim=(-30, 160))
plt.show()
```

```python
#Thematic edits
```

```python
list1 = []
mylabels = []
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)

sns.set_style('whitegrid')
fig, ax=plt.subplots()
fig.set_size_inches(11.0, 8.0)
plt.hist(list1, bins=30, stacked=True, label=mylabels) #add a list of colors to override -> colors=[]
plt.title("Movie Budget Distribution", fontsize=35, \
         color="DarkBlue", fontname="Console")
plt.ylabel("Number of Movies", fontsize=25, color="Red")
plt.xlabel("Budget", fontsize=25, color="Green")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xticks(np.arange(0, 320, 50.0))
plt.legend(frameon=True, fancybox= True, \
           shadow=True, framealpha=1, prop={'size':20})
plt.show()
```
